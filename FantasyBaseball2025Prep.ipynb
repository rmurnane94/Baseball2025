```python
import pandas
```

```python
ATC_projections = pandas.read_csv('ATC_projections.csv')
ATC_values = pandas.read_csv('ATC-auction-values.csv')
ATC_projections_simplified = ATC_projections.drop(columns = ['Name','G', 'AB', 'H', '1B', '2B', '3B', 'BB', 'SO', 'HBP', 'SF', 'SH', 'GDP', 'CS', 'BB%', 'IBB', 'K%', 'BB/K', 'ISO', 'BABIP', 'AVG', 'InterSD', 'InterSK', 'IntraSD', 'Spd', 'UBR', 'wSB', 'wRC', 'wRAA', 'BsR', 'Fld', 'Off', 'Def', 'WAR', 
                                                            'wRC+', 'Team', 'ADP'])
ATC_values_simplified = ATC_values.drop(columns = ['Team', 'ADP', 'PA', 'mRBI', 'mR', 'mSB', 'mHR', 'mOBP', 'mSLG', 'PTS', 'aPOS'])
joined_ATC = ATC_values_simplified.merge(ATC_projections_simplified, left_on='PlayerId', right_on='PlayerId')
```

```python
final_ATC = joined_ATC[['Name', 'POS', 'Dollars', 'PA', 'HR', 'R', 'RBI', 'SB', 'OBP', 'SLG', 'OPS','wOBA']]
final_ATC.sort_values(by=['Dollars'])
final_ATC = final_ATC.head(300)
final_ATC
```

```python
final_ATC.to_csv('RyanBatters2024.csv')
```

```python
ATC_pitcher_projections = pandas.read_csv('ATC_pitcherprojections.csv')
ATC_pitcher_projections_simplified = ATC_pitcher_projections.drop(columns = ['Team', 'W', 'L', 'G', 'GS', 'TBF', 'H', 'R', 'ER', 'HR', 'BB', 'IBB', 'HBP', 'K/BB', 'HR/9', 'K%', 'BB%','AVG', 'BABIP', 'LOB%',
       'GB%', 'HR/FB', 'FIP', 'WAR', 'RA9-WAR',  'InterSD', 'InterSK',
       'IntraSD',])
ATC_pitcher_values = pandas.read_csv('ATC_pitchervalues.csv')
ATC_pitcher_values_simplified = ATC_pitcher_values.drop(columns = ['Name','Team','ADP', 'IP', 'mSV', 'mERA', 'mWHIP', 'mSO',
       'mHLD', 'mQS', 'PTS', 'aPOS'])
joined_ATC_pitchers = ATC_pitcher_values_simplified.merge(ATC_pitcher_projections_simplified, left_on='PlayerId', right_on='PlayerId')
```

```python
final_ATC_pitchers = joined_ATC_pitchers[['Name', 'POS', 'Dollars', 'IP','ERA', 'WHIP',  'K/9', 'QS','SV', 'HLD', 'K-BB%']]
final_ATC_pitchers.sort_values(by=['Dollars'])

final_ATC_pitchers = final_ATC_pitchers.head(300)
final_ATC_pitchers
```

```python
PL_rankings = pandas.read_csv('PitcherList Rankings 2024.csv', encoding='latin-1')
Eno_ranks = pandas.read_csv('EnoRanks.csv')
```

```python
updated_pitchers =final_ATC_pitchers.merge(PL_rankings, how = 'left', left_on= 'Name', right_on= 'Name')
updated_pitchers = updated_pitchers.merge(Eno_ranks, how = 'left', left_on= 'Name', right_on= 'Name')
updated_pitchers.head(50)
```

```python
updated_pitchers.to_csv('RyanPitchers2024.csv')
```
