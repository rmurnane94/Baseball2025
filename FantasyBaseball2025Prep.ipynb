```python
import pandas
# Data/ATC_batter_projections.csv
```

```python
ATC_batter_projections = pandas.read_csv('Data/ATC_batter_projections.csv')
ATC_batter_values = pandas.read_csv('Data/ATC_batter_values.csv')
ATC_batter_projections_simplified = ATC_batter_projections.drop(columns = ['Name','G', 'AB', 'H', '1B', '2B', '3B', 'BB', 'SO', 'HBP', 'SF', 'SH', 'GDP', 'CS', 'BB%', 'IBB', 'K%', 'BB/K', 'ISO', 'BABIP', 'AVG', 'InterSD', 'InterSK', 'IntraSD', 'Spd', 'UBR', 'wSB', 'wRC', 'wRAA', 'BsR', 'Fld', 'Off', 'Def', 'WAR', 
                                                            'wRC+', 'Team', 'ADP'])
ATC_batter_values_simplified = ATC_batter_values.drop(columns = ['Team', 'ADP', 'PA', 'mRBI', 'mR', 'mSB', 'mHR', 'mOBP', 'mSLG', 'PTS', 'aPOS'])
joined_ATC = ATC_batter_values_simplified.merge(ATC_batter_projections_simplified, left_on='PlayerId', right_on='PlayerId')
```

```python
BatX_batter_projections = pandas.read_csv('Data/BatX_batter_projections.csv')
BatX_batter_values = pandas.read_csv('Data/BatX_batter_values.csv')
BatX_batter_projections_simplified = BatX_batter_projections.drop(columns = ['Name','G', 'AB', 'H', '1B', '2B', '3B', 'BB', 'SO', 'HBP', 'SF', 'SH', 'GDP', 'CS', 'BB%', 'IBB', 'K%', 'BB/K', 'ISO', 'BABIP', 'AVG', 'InterSD', 'InterSK', 'IntraSD', 'Spd', 'UBR', 'wSB', 'wRC', 'wRAA', 'BsR', 'Fld', 'Off', 'Def', 'WAR', 
                                                            'wRC+', 'Team', 'ADP'])
BatX_batter_values_simplified = BatX_batter_values.drop(columns = ['Team', 'ADP', 'PA', 'mRBI', 'mR', 'mSB', 'mHR', 'mOBP', 'mSLG', 'PTS', 'aPOS'])
joined_BatX = BatX_batter_values_simplified.merge(BatX_batter_projections_simplified, left_on='PlayerId', right_on='PlayerId')
```

```python
fantrax_hitters = pandas.read_csv('Data/Fantrax_hitters.csv')
fantrax_hitters = fantrax_hitters[['Player', 'ADP']]
fantrax_hitters['Fantrax_Rank'] = range(1, len(fantrax_hitters) + 1)

# hitters_fantrax_ranks = joined_ATC.merge(fantrax_hitters, left_on='NameASCII_y', right_on='Player').rename(columns={'ADP': 'Fantrax_ADP'})
hitters_fantrax_ranks = joined_BatX.merge(fantrax_hitters, left_on='NameASCII_y', right_on='Player').rename(columns={'ADP': 'Fantrax_ADP'})
```

```python
hitters_fantrax_ranks.sort_values('Fantrax_Rank').head(10)
```

```python
updated_hitters = hitters_fantrax_ranks[['Name', 'POS', 'Fantrax_Rank', 'Fantrax_ADP', 'Dollars', 'PA', 'HR', 'R', 'RBI', 'SB', 'OBP', 'SLG', 'OPS','wOBA']]
updated_hitters.sort_values(by=['Fantrax_Rank']).head(20)

```

```python
final_hitters = updated_hitters
```

```python
final_hitters.to_csv('RyanBattersBatX2025.csv')
# final_ATC.to_csv('RyanBatters2025.csv')
```

```python
ATC_pitcher_projections = pandas.read_csv('Data/ATC_pitcher_projections.csv')
ATC_pitcher_projections_simplified = ATC_pitcher_projections.drop(columns = ['Team', 'W', 'L', 'G', 'GS', 'TBF', 'H', 'R', 'ER', 'HR', 'BB', 'IBB', 'HBP', 'K/BB', 'HR/9', 'K%', 'BB%','AVG', 'BABIP', 'LOB%',
       'GB%', 'HR/FB', 'FIP', 'WAR', 'RA9-WAR',  'InterSD', 'InterSK',
       'IntraSD',])
ATC_pitcher_values = pandas.read_csv('Data/ATC_pitcher_values.csv')
ATC_pitcher_values_simplified = ATC_pitcher_values.drop(columns = ['Name','Team','ADP', 'IP', 'mSV', 'mERA', 'mWHIP', 'mSO',
        'mQS', 'PTS', 'aPOS'])
joined_ATC_pitchers = ATC_pitcher_values_simplified.merge(ATC_pitcher_projections_simplified, left_on='PlayerId', right_on='PlayerId')
```

```python
joined_ATC_pitchers.head()
```

```python
fantrax_pitchers = pandas.read_csv('Data/Fantrax_pitchers.csv')
fantrax_pitchers = fantrax_pitchers[['Player', 'ADP']]
fantrax_pitchers['Fantrax_Rank'] = range(1, len(fantrax_pitchers) + 1)

pitchers_fantrax_ranks = joined_ATC_pitchers.merge(fantrax_pitchers, left_on='NameASCII_y', right_on='Player').rename(columns={'ADP_y': 'Fantrax_ADP'})
```

```python
updated_pitchers = pitchers_fantrax_ranks[['Name', 'Fantrax_Rank', 'Fantrax_ADP', 'Dollars', 'IP','ERA', 'WHIP',  'K/9', 'QS','SV','K-BB%']]
updated_pitchers.sort_values(by=['Fantrax_Rank'], ascending=True).head(10)
```
